```python
import joblib
import numpy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sms
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
```

```python

gold_data.isnull().sum()
```

```python
x=gold_data.drop(['Date','GLD'],axis=1)
y=gold_data['GLD']
from sklearn.preprocessing import LabelEncoder
label_en_y=LabelEncoder()
y=label_en_y.fit_transform(y)
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=2)
fore=RandomForestClassifier(random_state=2,criterion="entropy",n_estimators=100)

fore.fit(X_train,Y_train)
forest=RandomForestRegressor(n_estimators=100)
forest.fit(X_train,Y_train)
```

```python
prd=forest.predict(X_test)[0]
prd1=fore.predict(X_test)[2]
print(prd)
print("Random Classifier",prd1)

from joblib import dump
dump(fore,"/content/drive/MyDrive/Colab Notebooks/goldprice_md.joblib")


```
